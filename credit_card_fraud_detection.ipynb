```python
X = df.drop("Class", axis=1)
y = df["Class"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=15,
    stratify=y
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("\ny_train distribution:")
print(y_train.value_counts(normalize=True) * 100)
print("\ny_test distribution:")
print(y_test.value_counts(normalize=True) * 100)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = X_train.copy()
X_test = X_test.copy()

X_train[["Time", "Amount"]] = scaler.fit_transform(X_train[["Time", "Amount"]])
X_test[["Time", "Amount"]] = scaler.transform(X_test[["Time", "Amount"]])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

log_reg = LogisticRegression(
    class_weight="balanced",
    random_state=15,
    max_iter=1000
)

log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)[:, 1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_prob))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
baseline_results = pd.DataFrame({
    "Model": ["Logistic Regression"],
    "Accuracy": [accuracy_score(y_test, y_pred)],
    "Precision": [precision_score(y_test, y_pred, zero_division=0)],
    "Recall": [recall_score(y_test, y_pred, zero_division=0)],
    "F1-Score": [f1_score(y_test, y_pred, zero_division=0)],
    "ROC-AUC": [roc_auc_score(y_test, y_prob)]
})

baseline_results
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=15,
    class_weight="balanced",
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_prob_rf))
```

Final Conclusion

The primary objective of this project was to detect any fraudulent credit card transactions in a heavily imbalanced dataset. Once a preliminary analysis was carried out, the duplicate entries were filtered out, and then the dataset was separated by stratified sampling technique that would keep the original proportions intact. Because of the presence of class imbalance, the performance of the models was measured in terms of Precision, Recall, F1 Score, and ROC-AUC instead of Accuracy.

Logistic Regression is a basic machine learning model and exhibited great Recall, but still produced a lot of false positives. Random Forest had much better Precision but could not detect many fraud cases. On the other hand, XGBoost and Light GBM showed balanced results in both terms, as they were able to detect frauds while at the same time not producing too many false positives. Finally, Tuned XGBoost gave the best results in all respects.
